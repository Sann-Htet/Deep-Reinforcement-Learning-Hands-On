```python
import numpy as np
import ptan

q_vals = np.array([[1, 2, 3], [1, 1, 0]])
q_vals
```

```python
selector = ptan.actions.ArgmaxActionSelector()
selector(q_vals)
```

```python
selector = ptan.actions.EpsilonGreedyActionSelector(epsilon=0.0, selector=ptan.actions.ArgmaxActionSelector())
selector(q_vals)
```

```python
selector = ptan.actions.EpsilonGreedyActionSelector(epsilon=1.0, selector=ptan.actions.ArgmaxActionSelector())
selector(q_vals)
```

```python
selector.epsilon
```

```python
selector.epsilon = 0.0
selector(q_vals)
```

```python
selector = ptan.actions.ProbabilityActionSelector()
for _ in range(10):
    acts = selector(np.array([
        [0.1, 0.8, 0.1],
        [0.0, 0.0, 1.0],
        [0.5, 0.5, 0.0]
    ]))
    print(acts)
```
